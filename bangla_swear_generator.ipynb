```python
from bltk.langtools import remove_stopwords
from bltk.langtools import Tokenizer
from bangla_stemmer.stemmer import stemmer

tokenizer_bltk = Tokenizer()
stmr = stemmer.BanglaStemmer()

bengali_punctuation = ['।', '॥', ',', '!', '?', ':', ';', '(', ')', '[', ']', '{', '}', '‘', '’', '“', '”', '–', '-', '—', '…', '😂', '😁', '🤣', '😡', '🤔', '😍', '👍', '😭', '😎','👞','👟','👠','👡','👢']

xLine =[]
for line in X:
  tokened_words = tokenizer_bltk.word_tokenizer(line)
  # stem_words = stemmer.stem(tokened_words)
  stp_words = remove_stopwords(tokened_words)
  stm = stmr.stem(stp_words)
  if len(stp_words) <= 10:
    sentence = ' '.join(stm)
    xLine.append(sentence)
```

```python
print(xLine[:11])
p=[]
for l in xLine:
  w = []
  for i in l:
    if i not in bengali_punctuation:
      w.append(i)
  sen = ''.join(w)
  p.append(sen)
print(p[:11])
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(lower=True)
tokenizer.fit_on_texts(p)
```

```python
x_sequence = tokenizer.texts_to_sequences(p)
```

```python
print(tokenizer.word_index.items())
print(len(tokenizer.word_index))


```

```python
x_train = []
y_train = []
for line in x_sequence:
  for i in range(1,len(line)-1):
    x_train.append(line[:i+1])
    y_train.append(line[i+1])
```

```python
vocab_size = len(tokenizer.word_index)+1
word2idx = tokenizer.word_index.items()
```

```python
print(x_train[:10])
print(y_train[:10])
print(len(x_train))
print(len(y_train))
```

```python
max_seq_length = max(len(x) for x in x_train)
print(max_seq_length)
```

```python
max_length = max_seq_length
trunc_type='pre'
padding_type='pre'
x_train_padded = pad_sequences(x_train, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
x_train_padded[0]
```

```python
from tensorflow.keras.utils import to_categorical
y_lable = to_categorical(y_train, num_classes=vocab_size)
print(len(y_lable))
```

```python
y_lable.shape
```

```python
x_Train = x_train_padded[:20000]
x_Test =x_train_padded[20000:25000]

y_Train = y_lable[:20000]
y_Test = y_lable[20000:25000]
```

```python
from tensorflow.keras.layers import Input, Dense, Embedding, LSTM, Dropout, Bidirectional, GlobalMaxPooling1D, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Embedding(vocab_size, 256, input_length=max_seq_length))
model.add(Dropout(0.5))
model.add(LSTM(500, return_sequences=True))
model.add(Bidirectional(LSTM(1000, return_sequences=True)))
model.add(LSTM(500, return_sequences=False))
model.add(Dense(8000, activation='relu'))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(optimizer='adam',
              loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
print(x_Train.shape)
print(y_Train.shape)
```

```python
model.fit(x_Train,y_Train, batch_size=128, epochs=20,validation_split=0.15)
```
